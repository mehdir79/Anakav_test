```python
import json
import pandas as pd
import ast  # امن‌تر از eval

with open("q.json", encoding="utf-8") as f:
    raw_str = json.load(f)  # این هنوز str هست
    data = json.loads(raw_str)  # حالا تبدیل به dict واقعی میشه

# حالا تبدیل کلیدها به tuple
data_fixed = {
    ast.literal_eval(k): v
    for k, v in data.items()
}

df = pd.DataFrame(data_fixed)
df.columns = pd.MultiIndex.from_tuples(df.columns)
df

```
